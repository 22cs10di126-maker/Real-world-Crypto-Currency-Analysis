```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# -----------------------------------
# SETTINGS
# -----------------------------------

headers = {
    "User-Agent":"Mozilla/5.0"
}

crypto_data=[]


# -----------------------------------
# SCRAPE 20 PAGES
# -----------------------------------

for page in range(1,21):

    print(f"Scraping page {page}...")

    url=f"https://coinmarketcap.com/?page={page}"

    try:

        response=requests.get(
            url,
            headers=headers,
            timeout=15
        )

        if response.status_code!=200:
            print(f"Skipping page {page}")
            continue


        soup=BeautifulSoup(
            response.content,
            "html.parser"
        )

        rows=soup.find_all("tr")

        records=0


        for row in rows:

            tds=row.find_all("td")

            if len(tds)>9:

                data={}

                # -------------------
                # RANK FILTER
                # removes junk rows
                # -------------------

                rank=tds[1].get_text(
                    strip=True
                )

                if not rank.isdigit():
                    continue


                # -------------------
                # COIN NAME + SYMBOL
                # -------------------

                p_tags=tds[2].find_all("p")

                coin_name=(
                    p_tags[0].get_text(strip=True)
                    if len(p_tags)>0
                    else None
                )

                symbol=(
                    p_tags[1].get_text(strip=True)
                    if len(p_tags)>1
                    else None
                )


                if not coin_name or not symbol:
                    continue


                # -------------------
                # REMOVE NON-COINS
                # -------------------

                bad_words=[
                    "Index",
                    "DTF",
                    "CoinMarketCap",
                    "CMC",
                    "Portfolio"
                ]

                if any(
                    word in coin_name
                    for word in bad_words
                ):
                    continue


                if not symbol.isupper():
                    continue


                if len(symbol)>8:
                    continue


                data["Rank"]=rank
                data["Coin Name"]=coin_name
                data["Symbol"]=symbol



                # -------------------
                # PRICE RAW
                # -------------------

                price_tag=tds[3].find(
                    "span"
                )

                raw_price=(
                    price_tag.get_text(
                        strip=True
                    )
                    if price_tag
                    else None
                )

                data["Price (USD)"]=raw_price



                # -------------------
                # 1H 24H 7D RAW
                # -------------------

                c1=tds[4].find("span")
                c24=tds[5].find("span")
                c7=tds[6].find("span")


                ch1=(
                    c1.get_text(strip=True)
                    if c1 else None
                )

                ch24=(
                    c24.get_text(strip=True)
                    if c24 else None
                )

                ch7=(
                    c7.get_text(strip=True)
                    if c7 else None
                )


                data["1h Change (%)"]=ch1
                data["24h Change (%)"]=ch24
                data["7d Change (%)"]=ch7



                # -------------------
                # MARKET CAP
                # -------------------

                mc_spans=tds[7].find_all(
                    "span"
                )

                if len(mc_spans)>=2:
                    data["Market Cap"]=(
                        mc_spans[1].get_text(
                            strip=True
                        )
                    )

                elif len(mc_spans)==1:
                    data["Market Cap"]=(
                        mc_spans[0].get_text(
                            strip=True
                        )
                    )

                else:
                    data["Market Cap"]=None



                # -------------------
                # VOLUME
                # -------------------

                vol_ps=tds[8].find_all(
                    "p"
                )

                if len(vol_ps)>0:
                    data[
                    "Volume (24h) USD"
                    ]=(
                        vol_ps[0].get_text(
                            strip=True
                        )
                    )
                else:
                    data[
                    "Volume (24h) USD"
                    ]=None



                # -------------------
                # CIRCULATING SUPPLY
                # -------------------

                supply_span=tds[9].find(
                    "span"
                )

                data[
                "Circulating Supply"
                ]=(
                    supply_span.get_text(
                        strip=True
                    )
                    if supply_span
                    else None
                )



                # -------------------
                # AVG DOWNFALL
                # -------------------

                try:

                    p1=float(
                        ch1.replace("%","")
                    )

                    p24=float(
                        ch24.replace("%","")
                    )

                    p7=float(
                        ch7.replace("%","")
                    )

                    avg=(
                        p1+p24+p7
                    )/3

                except:
                    avg=None


                data[
                "Average Downfall %"
                ]=avg



                # -------------------
                # PRICE RANGE
                # -------------------

                try:

                    price_num=float(
                        raw_price
                        .replace("$","")
                        .replace(",","")
                    )

                    if price_num<=0.05:
                        bucket="$0-$0.05"

                    elif price_num<=0.5:
                        bucket="$0.05-$0.5"

                    elif price_num<=5:
                        bucket="$0.5-$5"

                    elif price_num<=50:
                        bucket="$5-$50"

                    else:
                        bucket=">$50"

                except:
                    bucket=None


                data["Price Range"]=bucket


                crypto_data.append(
                    data
                )

                records+=1



        print(
            f"Records scraped: {records}"
        )

        time.sleep(2)


    except Exception as e:

        print(
            f"Error on page {page}: {e}"
        )



# -----------------------------------
# EXPORT RAW DATA
# -----------------------------------

df=pd.DataFrame(
    crypto_data
)

print(
f"\nTotal rows scraped: {len(df)}"
)


file_name="analysis_p1_raw.xlsx"

df.to_excel(
    file_name,
    index=False
)

print(
f"\nExcel exported: {file_name}"
)

print("\nPreview:")
print(
df.head(10)
)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# -----------------------------------
# SETTINGS
# -----------------------------------

headers = {
    "User-Agent": "Mozilla/5.0"
}

crypto_data = []

# -----------------------------------
# WORDS TO REMOVE
# -----------------------------------

bad_words = [
    "stock",
    "tokenized",
    "ETF",
    "Index",
    "Portfolio",
    "CoinMarketCap",
    "CMC"
]

# -----------------------------------
# SCRAPE PAGES
# -----------------------------------

for page in range(1, 21):

    print(f"\nScraping Page {page}...")

    url = f"https://coinmarketcap.com/?page={page}"

    try:

        response = requests.get(
            url,
            headers=headers,
            timeout=15
        )

        if response.status_code != 200:
            print(f"Skipping Page {page}")
            continue

        soup = BeautifulSoup(
            response.content,
            "html.parser"
        )

        rows = soup.find_all("tr")

        records = 0

        for row in rows:

            tds = row.find_all("td")

            if len(tds) > 9:

                # -----------------------------------
                # VALID RANK ONLY
                # -----------------------------------

                rank = tds[1].get_text(strip=True)

                if not rank.isdigit():
                    continue

                # -----------------------------------
                # COIN NAME + SYMBOL
                # -----------------------------------

                p_tags = tds[2].find_all("p")

                coin_name = (
                    p_tags[0].get_text(strip=True)
                    if len(p_tags) > 0
                    else None
                )

                symbol = (
                    p_tags[1].get_text(strip=True)
                    if len(p_tags) > 1
                    else None
                )

                if not coin_name or not symbol:
                    continue

                # -----------------------------------
                # REMOVE NON-CRYPTO ROWS
                # -----------------------------------

                coin_lower = coin_name.lower()

                if any(
                    word.lower() in coin_lower
                    for word in bad_words
                ):
                    continue

                # -----------------------------------
                # PRICE
                # -----------------------------------

                price_tag = tds[3].find("span")

                raw_price = (
                    price_tag.get_text(strip=True)
                    if price_tag
                    else None
                )

                # -----------------------------------
                # 1H, 24H, 7D %
                # -----------------------------------

                c1 = tds[4].find("span")
                c24 = tds[5].find("span")
                c7 = tds[6].find("span")

                ch1 = (
                    c1.get_text(strip=True)
                    if c1 else None
                )

                ch24 = (
                    c24.get_text(strip=True)
                    if c24 else None
                )

                ch7 = (
                    c7.get_text(strip=True)
                    if c7 else None
                )

                # -----------------------------------
                # STORE DATA
                # -----------------------------------

                data = {
                    "Rank": rank,
                    "Coin Name": coin_name,
                    "Symbol": symbol,
                    "Price (USD)": raw_price,
                    "1h Change (%)": ch1,
                    "24h Change (%)": ch24,
                    "7d Change (%)": ch7
                }

                crypto_data.append(data)

                records += 1

        print(f"Records Scraped: {records}")

        time.sleep(2)

    except Exception as e:

        print(f"Error on Page {page}: {e}")

# -----------------------------------
# CREATE DATAFRAME
# -----------------------------------

df = pd.DataFrame(crypto_data)

# -----------------------------------
# REMOVE DUPLICATES
# -----------------------------------

df = df.drop_duplicates(
    subset=["Coin Name"]
)

# -----------------------------------
# EXPORT EXCEL
# -----------------------------------

file_name = "crypto_clean_raw_data.xlsx"

df.to_excel(
    file_name,
    index=False
)

print(f"\nExcel Exported: {file_name}")

print("\nPreview:\n")

print(df.head(10))
```
